## Tarea 2: Queries con SQL

### Ejercicio 1

##### Muestra las diferentes plataformas de lanzamiento:

```python
#Rellena la variable query con tu consulta
query = """select LaunchSite from SPACEXTBL;"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
#Rellena la variable query con tu consulta
query = """SELECT LaunchSite	FROM SPACEXTBL WHERE LaunchSite LIKE 'CCS%' ORDER BY LaunchSite ASC LIMIT 5  """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
query = """ SELECT Customers, PayloadMass FROM SPACEXTBL WHERE Customers LIKE '%NASA (CRS)%' ORDER BY PayloadMass DESC """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
query = """ SELECT Serial, PayloadMass FROM SPACEXTBL WHERE Serial LIKE 'B1%' ORDER BY PayloadMass DESC  """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
query = """  SELECT MIN(Date) AS Fecha, LandingOutcome FROM SPACEXTBL WHERE LandingOutcome LIKE '%TRUE RTLS%'  LIMIT 1   """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
query = """  SELECT BoosterVersion AS Cohetes, LandingOutcome, PayloadMass FROM SPACEXTBL WHERE LandingOutcome LIKE '%TRUE ASDS%' AND PayloadMass  BETWEEN 4000 AND 6000 ORDER BY PayloadMass DESC    """
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
query = """SELECT COUNT (case when LandingOutcome LIKE '%TRUE%' then LandingOutcome END) Extitoso
    COUNT (case when LandingOutcome LIKE '%FALSE%' then LandingOutcome END) as  Fracaso,
    FROM SPACEXTBL
    WHERE LandingOutcome != 'None'
    GROUP BY LandingOutcome"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
query = """SELECT BoosterVersion, PayloadMass, LaunchSite, COUNT(*) AS Cantidad
    FROM SPACEXTBL
    GROUP BY PayloadMass"""
df  = pd.read_sql(query, con = engine )
df
```

### Ejercicio 9

##### Muestra los el resultados del aterrizaje, la serie del cohete y las plataforma de lanzamiento de aquellos lanzamientos fallidos o no intentados en el año 2015


```python
query = """  SELECT Date, BoosterVersion, LandingOutcome FROM SPACEXTBL WHERE LandingOutcome LIKE '%FALSE%' AND Date = 2015 ORDER BY BoosterVersion DESC"""
df  = pd.read_sql(query, con = engine )
df
```
